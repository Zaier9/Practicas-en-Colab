```python
import matplotlib.pyplot as plt
import math
from dataclasses import dataclass
```

```python
@dataclass
class MakeSpiral:
    radius: float=35
    num_turns: int=5
    point_per_turns: int=60
    special_twist: bool=False

    def plot(self):
        AllPointsToPlot=360*self.num_turns
        DegreeDeltaPerPoint=int(360/self.point_per_turns)

        fig=plt.figure()
        ax=fig.add_subplot(111, projection='3d')
        X=[]
        Y=[]
        Z=[]
        for i in range(0, AllPointsToPlot, DegreeDeltaPerPoint):
            theta=i/360*2*math.pi
            coefficient=1
            if self.special_twist:
                coefficient = i/AllPointsToPlot

            x = self.radius*coefficient*math.sin(theta)

            y = self.radius*coefficient*math.cos(theta)

            z = i
            X.append(x)
            Y.append(y)
            Z.append(z)

        ax.scatter3D(X, Y, Z)
        plt.savefig('spiral.png')
        plt.show()

if __name__ == '__main__':
    spiral=MakeSpiral()
    spiral.plot()
```
